### Exercise 1

The goal of lesson 10 is to *adapt* the Genetic Algorithm code written in the previous lesson to have a C++ code to solve the TSP with a Simulated Annealing (SA) algorithm.
The program exploits the previously implemented genetic mutation (randomly drawn) as proposed move for the algorithm.

Like in the previous lesson I applied my code to optimize the path among 30 cities randomly placed on a circumference and 30 cities randomly placed inside a square.

Once I obtained the data I produced a picture the length of the optimized path as a function of the iterations of the SA algorithm and a picture of such path for both the circular and the square dispositions of the cities.

Chosen parameters are:

- maximum (start) temperature $T_{max}=1.5$
- minimum (end) temperature $T_{min}=0.001$
- temperature steps: $1000$
- moves per tempeature: $100$

Results are shown below.

```python
x_sq, y_sq = np.loadtxt("Exercise_01/data/square/output_distance.dat", unpack = True)
x_sq_path, y_sq_path = np.loadtxt("Exercise_01/data/square/best_path.dat", unpack = True)

x_circ, y_circ = np.loadtxt("Exercise_01/data/circle/output_distance.dat", unpack = True)
x_circ_path, y_circ_path = np.loadtxt("Exercise_01/data/circle/best_path.dat", unpack = True)

plt.figure(1, figsize = (9, 6))

plt.title("Lenght of the path (square world)")
plt.plot(x_sq, y_sq)
plt.xlabel('Iterations')
plt.ylabel('$L^{(1)}$')
plt.grid(True)

plt.figure(2, figsize = (9, 6))

plt.title("Lenght of the path (circular world)")
plt.plot(x_circ, y_circ)
plt.xlabel('Iterations')
plt.ylabel('$L^{(1)}$')
plt.grid(True)

plt.figure(3, figsize = (13, 6))

plt.subplot(1, 2, 1)

plt.title("Best path (square world)")
plt.plot(x_sq_path, y_sq_path,  marker=".")
plt.axis('equal')
plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.xticks([])
plt.yticks([])

plt.subplot(1, 2, 2)

plt.title("Best path (circular world)")
plt.plot(x_circ_path, y_circ_path,  marker=".")
plt.axis('equal')
plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.xticks([])
plt.yticks([])
```

### Exercise 2

For this exercise I implemented the possibility to parallelize the SA research of the optimal path using MPI libraries.
The implementation of parallelization is very naive: in my code each *node* performs an independent SA search (starting from different paths) and only in the end indipendent results are compared, and the best one is chosen to be the output of the program.

I applied the code to the exact same TSP problem shown above, requiring 4 nodes of calculus.

Results are the same obtained with the non-parallel code and are reported below.

```python
x_sq_0, y_sq_0 = np.loadtxt("Exercise_02/data/square/output_distance0.dat", unpack = True)
x_sq_1, y_sq_1 = np.loadtxt("Exercise_02/data/square/output_distance1.dat", unpack = True)
x_sq_2, y_sq_2 = np.loadtxt("Exercise_02/data/square/output_distance2.dat", unpack = True)
x_sq_3, y_sq_3 = np.loadtxt("Exercise_02/data/square/output_distance3.dat", unpack = True)
x_sq_path_mpi, y_sq_path_mpi = np.loadtxt("Exercise_02/data/square/best_path.dat", unpack = True)

x_circ_0, y_circ_0 = np.loadtxt("Exercise_02/data/circle/output_distance0.dat", unpack = True)
x_circ_1, y_circ_1 = np.loadtxt("Exercise_02/data/circle/output_distance1.dat", unpack = True)
x_circ_2, y_circ_2 = np.loadtxt("Exercise_02/data/circle/output_distance2.dat", unpack = True)
x_circ_3, y_circ_3 = np.loadtxt("Exercise_02/data/circle/output_distance3.dat", unpack = True)
x_circ_path_mpi, y_circ_path_mpi = np.loadtxt("Exercise_02/data/circle/best_path.dat", unpack = True)

plt.figure(1, figsize = (9, 6))

plt.title("Lenght of the path (square world)")
plt.plot(x_sq_0, y_sq_0, label="Node 0")
plt.plot(x_sq_1, y_sq_1, label="Node 1")
plt.plot(x_sq_2, y_sq_2, label="Node 2")
plt.plot(x_sq_3, y_sq_3, label="Node 3")
plt.xlabel('Iterations')
plt.ylabel('$L^{(1)}$')
plt.grid(True)
plt.legend()

plt.figure(2, figsize = (9, 6))

plt.title("Lenght of the path (circular world)")
plt.plot(x_circ_0, y_circ_0, label="Node 0")
plt.plot(x_circ_1, y_circ_1, label="Node 1")
plt.plot(x_circ_2, y_circ_2, label="Node 2")
plt.plot(x_circ_3, y_circ_3, label="Node 3")
plt.xlabel('Iterations')
plt.ylabel('$L^{(1)}$')
plt.grid(True)
plt.legend()

plt.figure(3, figsize = (13, 6))

plt.subplot(1, 2, 1)

plt.title("Best path (square world)")
plt.plot(x_sq_path_mpi, y_sq_path_mpi,  marker=".")
plt.axis('equal')
plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.xticks([])
plt.yticks([])

plt.subplot(1, 2, 2)

plt.title("Best path (circular world)")
plt.plot(x_circ_path_mpi, y_circ_path_mpi,  marker=".")
plt.axis('equal')
plt.xlim((-1, 1))
plt.ylim((-1, 1))
plt.xticks([])
plt.yticks([])
```
