### Exercise 1

The purpose of this exercise is to test the Pseudo-Random Number generator downloaded from the NSL Ariel web site.
The first step consists in making an estimation of  $\langle r \rangle = \int_0^1 r dr = 1/2$ and its statistical uncertainty using the blocking method, then to compute the variance value in function of number of blocks and finally to perform a chi-squared test to show the goodness of the generator.

```python
plt.figure(1, figsize = (9, 6))
x_mean, y_mean, yerr_mean = np.loadtxt("Exercise_01/output01.1.dat", delimiter = "   ", unpack = True)
plt.title("Mean value")
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.axhline(y=0, color='r', linestyle="dotted", label = "Expected value")
plt.errorbar(x_mean, y_mean, yerr_mean, label = "Data")
plt.legend()

plt.figure(2, figsize = (9, 6))
x_sigma, y_sigma, yerr_sigma = np.loadtxt("Exercise_01/output01.2.dat", delimiter = "   ", unpack = True)
plt.title("Variance")
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.axhline(y=0, color='r', linestyle="dotted", label = "Expected value")
plt.errorbar(x_sigma, y_sigma, yerr_sigma, label = "Data")
plt.legend()

plt.figure(3, figsize = (9, 6))
x_chi = np.loadtxt("Exercise_01/output01.3.dat")
plt.plot(x_chi, label = "Data")
plt.title("Chi-squared test")
plt.xlabel("Blocks")
plt.ylabel(r"$\chi^2$")
plt.grid(True)
plt.axhline(y=100, color='r', linestyle="dotted", label = "Expected value")
plt.legend()
```

The plot of the mean value shows a discrepancy between the expected value and the data obtained greater than the standard deviation, although within twice its value.
The average value of the variance is within the error bar and in good agreement with the expected value.

The chi-squared test, calculated dividing $[0,1]$ into $M=100$ sub-intervals, revealed a value averaged over $j=1,...,100$ realizations of $n=10^4$ throws, of $\chi^2=98.8$, very close to the expected one ($\chi^2_{expected}=M=100$).
The last plot shows the $\chi^2$ value calculated in each realization. We can see it oscillating around the expected value, as a confirmation that we cannot say that our numbers are **not** drawn from a uniform distribution. We therefore assume them to be uniformly distributed.

### Exercise 2

The aim of this exercise is to extend the Pseudo-Random Number generator by adding two probability distributions using the method of the inversion of the cumulative distribution to sample an exponential distribution $p(x) = \lambda \exp(-\lambda x)$, $x\in [0;+\infty]$ and a Cauchy-Lorentz distribution $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$, $x\in [-\infty;+\infty]$.
Than we make 3 pictures, each of them containing 4 histograms obtained with $10^4$ realizations of $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ varying $N=1,2,10,100$, using the *uniform dice*, *exponential dice* and *Lorentzian dice* respectively.

```python
plt.figure(1, figsize = (9, 6))

x_std_1, y_std_1 = np.loadtxt("Exercise_02/data/hist_std_1.dat", delimiter = "  ", unpack = True)
x_std_2, y_std_2 = np.loadtxt("Exercise_02/data/hist_std_2.dat", delimiter = "  ", unpack = True)
x_std_10, y_std_10 = np.loadtxt("Exercise_02/data/hist_std_10.dat", delimiter = "  ", unpack = True)
x_std_100, y_std_100 = np.loadtxt("Exercise_02/data/hist_std_100.dat", delimiter = "  ", unpack = True)
plt.step(x_std_1, y_std_1, where='mid', color="tab:red", label="N=1")
plt.step(x_std_2, y_std_2, where='mid', color="tab:green", label="N=2")
plt.step(x_std_10, y_std_10, where='mid', color="tab:blue", label="N=10")
plt.step(x_std_100, y_std_100, where='mid', color="tab:orange", label="N=100")
plt.legend()
plt.grid(True)
plt.title("Draws from uniform distribution")

plt.figure(2, figsize = (9, 6))

x_exp_1, y_exp_1 = np.loadtxt("Exercise_02/data/hist_exp_1.dat", delimiter = "  ", unpack = True)
x_exp_2, y_exp_2 = np.loadtxt("Exercise_02/data/hist_exp_2.dat", delimiter = "  ", unpack = True)
x_exp_10, y_exp_10 = np.loadtxt("Exercise_02/data/hist_exp_10.dat", delimiter = "  ", unpack = True)
x_exp_100, y_exp_100 = np.loadtxt("Exercise_02/data/hist_exp_100.dat", delimiter = "  ", unpack = True)
plt.step(x_exp_1, y_exp_1, where='mid', color="tab:red", label="N=1")
plt.step(x_exp_2, y_exp_2, where='mid', color="tab:green", label="N=2")
plt.step(x_exp_10, y_exp_10, where='mid', color="tab:blue", label="N=10")
plt.step(x_exp_100, y_exp_100, where='mid', color="tab:orange", label="N=100")
plt.legend()
plt.grid(True)
plt.title("Draws from exponential distribution")

plt.figure(3, figsize = (9, 6))

x_lor_1, y_lor_1 = np.loadtxt("Exercise_02/data/hist_lor_1.dat", delimiter = "  ", unpack = True)
x_lor_2, y_lor_2 = np.loadtxt("Exercise_02/data/hist_lor_2.dat", delimiter = "  ", unpack = True)
x_lor_10, y_lor_10 = np.loadtxt("Exercise_02/data/hist_lor_10.dat", delimiter = "  ", unpack = True)
x_lor_100, y_lor_100 = np.loadtxt("Exercise_02/data/hist_lor_100.dat", delimiter = "  ", unpack = True)
plt.step(x_lor_1, y_lor_1, where='mid', color="tab:red", label="N=1")
plt.step(x_lor_2, y_lor_2, where='mid', color="tab:green", label="N=2")
plt.step(x_lor_10, y_lor_10, where='mid', color="tab:blue", label="N=10")
plt.step(x_lor_100, y_lor_100, where='mid', color="tab:orange", label="N=100")
plt.legend()
plt.grid(True)
plt.title("Draws from Cauchy-Lorentz distribution")
```

We observe that, increasing $N$, the first two distributions tend to became gaussian, since the central limit theorem applies.
In the latter case, since Cauchy-Lorentz distribution hasn't a finite variance, this fact doesn't happen.
However this distribution is stable, which means that increasing indefinitely $N$ we always get a Cauchy-Lorentz distribution.
We then try to fit the case $N=100$ with a Gaussian function for standard and exponential dices and a Lorentzian function for the last case.

```python
def gauss(x, mu, sigma):
    return 1. / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2 / (2 * sigma**2))
def lorentz(x, x0, y0):
    return 1 / np.pi * y0 / ((x - x0)**2 + y0**2)

popt_std, pcov_std = curve_fit(gauss, x_std_100, y_std_100)
popt_exp, pcov_exp = curve_fit(gauss, x_exp_100, y_exp_100)
popt_lor, pcov_lor = curve_fit(lorentz, x_lor_100, y_lor_100)

plt.figure(1, figsize = (9, 6))
plt.step(x_std_100, y_std_100, where='mid', color="tab:blue", label="Data")
plt.plot(np.arange(0.,1.,0.01), gauss(np.arange(0.,1.,0.01), popt_std[0], popt_std[1]), color="tab:red", linestyle="dotted", label="Fit")
plt.grid(True)
plt.legend()
plt.title("Draws from uniform distribution")

plt.figure(2, figsize = (9, 6))
plt.step(x_exp_100, y_exp_100, where='mid', color="tab:blue", label="Data")
plt.plot(np.arange(0.,5.,0.01), gauss(np.arange(0.,5.,0.01), popt_exp[0], popt_exp[1]), color="tab:red", linestyle="dotted", label="Fit")
plt.grid(True)
plt.legend()
plt.title("Draws from exponential distribution")

plt.figure(3, figsize = (9, 6))
plt.step(x_lor_100, y_lor_100, where='mid', color="tab:blue", label="Data")
plt.plot(np.arange(-10.,10.,0.01), lorentz(np.arange(-10.,10.,0.01), popt_lor[0], popt_lor[1]), color="tab:red", linestyle="dotted", label="Fit")
plt.grid(True)
plt.legend()
plt.title("Draws from Cauchy-Lorentz distribution")
```

In the first two cases a gaussian function seems to approximate well the distribution of the sum of the random variables, confirming the validity of central limit theorem.
In the third case we observe the self-similarity of the distribution of the sum of the Cauchy-Lorentz distributed random variables varying $N$ confirming the stability of this distribution.

### Exercise 3

In this exercise we wanted to simulate the Buffon’s experiment to calculate an estimate of the value of $\pi$.
The picture of the estimation of $\pi$ and its uncertainty, as a function of the number of blocks, is shown below.

```python
plt.figure(figsize = (9, 6))

x_pval, y_pval, yerr_pval = np.loadtxt("Exercise_03/output.dat", delimiter = "   ", unpack = True)
plt.title("Estimate of $\pi$")
plt.xlabel('Blocks')
plt.grid(True)
plt.axhline(y=np.pi, color='r', linestyle="dotted", label = "Expected value")
plt.errorbar(x_pval, y_pval, yerr_pval, label = "Data")
plt.legend()
```
